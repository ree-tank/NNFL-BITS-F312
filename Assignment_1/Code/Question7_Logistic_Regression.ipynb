```
import pandas as pd
import numpy as np
```

```
#read the data
data = pd.read_excel('data3.xlsx', header = None)
data.tail()
```

```
#change label 2 to label 0
data.loc[data[4]==2, 4] = 0
data.tail()
```

```
#shuffling the data
data = data.sample(frac = 1, random_state = 3).reset_index(drop = True)
data
```

```
#seperating input(X) and output(Y)
X = data.iloc[:,:-1]
Y = data.iloc[:,-1:]
print("First 5 rows of X")
print(X.head())
print()
print("First 5 rows of Y")
print(Y.head())
```

```
#splitting X and Y in test and training data
def split(X, Y, split_size = 0.6):
  train_size = int(len(X) * split_size)
  X_tr = X.iloc[:train_size, :]
  X_ts = X.iloc[train_size:, :]
  Y_tr = Y.iloc[:train_size, :]
  Y_ts = Y.iloc[train_size:, :]

  return X_tr, X_ts, Y_tr, Y_ts
```

```
X_train, X_test, Y_train, Y_test = split(X, Y, split_size = 0.6)
X_test = X_test.reset_index(drop = True)
Y_test = Y_test.reset_index(drop = True)
```

```
print("X_train shape is: ", X_train.shape)
print("Y_train shape is: ", Y_train.shape)
print("X_test shape is: ", X_test.shape)
print("Y_test shape is: ", Y_test.shape)
```

```
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```
def compute_cost(A, Y):
  cost_val = (Y*np.log(A)) + ((1-Y)*np.log(1-A))
  return cost_val 
```

```
c = compute_cost(0.91,1)
print(c)
```

```
def update_params(X, Y, w0, w1, w2, w3, w4, learning_rate = 0.01):
  dw0 = dw1 = dw2 = dw3 = dw4 = 0
  m = X.shape[0]
  A = [0 for i in range(len(X))]
  cost = 0

  for i in range(len(X)):
    #print("i is:", i)
    A[i] = float(sigmoid(w0 + w1*X.iloc[i,0] + w2*X.iloc[i,1] + w3*X.iloc[i,2] + w4*X.iloc[i,3]))
    #print(A[i])
    #print(type(Y[i]))
    cost = cost + compute_cost(A[i] , Y.iloc[i,0])
    #print("cost is:", end = " ")
    #print(cost)
    dw0 = dw0 + (A[i] - Y.iloc[i,0])
    #print(w0)
    dw1 = dw1 + (A[i] - Y.iloc[i,0])*X.iloc[i,0]
    #print(w1)
    dw2 = dw2 + (A[i] - Y.iloc[i,0])*X.iloc[i,1]
    dw3 = dw3 + (A[i] - Y.iloc[i,0])*X.iloc[i,2]
    dw3 = dw3 + (A[i] - Y.iloc[i,0])*X.iloc[i,3]

  cost = (-1/m)*cost
  w0 = w0 - learning_rate * (dw0/m)
  w1 = w1 - learning_rate * (dw1/m)
  w2 = w2 - learning_rate * (dw2/m)
  w3 = w3 - learning_rate * (dw3/m)
  w4 = w4 - learning_rate * (dw4/m)

  return w0,w1,w2,w3,w4,cost
```

```
def log_reg(X, Y, learning_rate = 0.01, num_iter = 1000):
  np.random.seed(3)
  w0 = float(np.random.randn(1,1))
  w1 = float(np.random.randn(1,1))
  w2 = float(np.random.randn(1,1))
  w3 = float(np.random.randn(1,1))
  w4 = float(np.random.randn(1,1))

  cost_list = []

  for i in range(num_iter):
    w0, w1, w2, w3, w4, cost = update_params(X, Y, w0, w1, w2, w3, w4, learning_rate = 0.01)
    cost_list.append(cost)

  return w0, w1, w2, w3, w4, cost, cost_list  
```

```
w0_f, w1_f, w2_f, w3_f, w4_f, cost_f, cost_list_f = log_reg(X_train, Y_train, learning_rate = 0.01, num_iter = 1000)
```

```
def pred(X, Y, w0, w1, w2, w3, w4):
  A_test = [0 for i in range(len(X))]
  Y_pred = [0 for i in range(len(X))]

  for i in range(len(X)):
    A_test[i] = float(sigmoid(w0 + w1*X.iloc[i,0] + w2*X.iloc[i,1] + w3*X.iloc[i,2] + w4*X.iloc[i,3]))
    Y_pred[i] = 1 if (A_test[i] >= 0.5) else 0

  Y_pred = pd.DataFrame(Y_pred)
  return Y_pred
```

```
Y_output = pred(X_test, Y_test, w0_f, w1_f, w2_f, w3_f, w4_f)
print(Y_output.head())
```

```
def confusion_matrix(Y, Yp):
  tp = tn = fp = fn = 0
  for i in range(len(Y)):
    if(Y.iloc[i,0]==1):
      if(Yp.iloc[i,0]==1):
        tp = tp + 1
      else:
        fn = fn + 1
    else:
      if(Yp.iloc[i,0]==1):
        fp = fp + 1
      else:
        tn = tn + 1
  return tp, fp, fn, tn
```

```
tp, fp, fn, tn = confusion_matrix(Y_test, Y_output)
print("True positive is: ", tp)
print("False positive is: ", fp)
print("False negetive is: ", fn)
print("True negetive is: ", tn)
```

```
accuracy = (tp+tn)/(tp+tn+fp+fn)
sensitivity = tp/(tp+fn)
specificity = tn/(tn+fp)

print("accuracy is: {0:13f}%".format(accuracy*100))
print("sensitivity is: {0:5f}%".format(sensitivity*100))
print("specificity is: {0:5f}%".format(specificity*100))
```
