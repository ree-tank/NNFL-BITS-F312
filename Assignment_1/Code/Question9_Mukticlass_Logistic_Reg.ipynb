```
import pandas as pd
import numpy as np
```

```
### read the data ####
data = pd.read_excel('data4.xlsx', header = None)
#data.head()
```

```
## shuffling the data ##
data_shuffled = data.sample(frac = 1, random_state = 7).reset_index(drop = True)
#data_shuffled.head()
```

```
## separate input(X) and output(Y)
X = data_shuffled.iloc[:,:-1]
Y = data_shuffled.iloc[:,-1:]
#print("First 5 rows of X is: ")
#print(X.head())
#print()
#print("First 5 rows of Y is: ")
Y.head(50)
```

```
## function to normalize data ##
def normalize(D, min_features, max_features):
  Z = pd.DataFrame()
  for i in range(D.shape[1]):
    Z[i] = (D.iloc[:,i] - min_features[i])/(max_features[i] - min_features[i])
  return Z
```

```
## sigmoid function
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```
## cost function
def compute_cost(A, Y):
  cost_val = (Y*np.log(A)) + ((1-Y)*np.log(1-A))
  return cost_val 
```

```
## update parameters
def update_params(X, Y, w0, w1, w2, w3, w4, w5, w6, w7, learning_rate = 0.01):
  dw0 = dw1 = dw2 = dw3 = dw4 = dw5 = dw6 = dw7 = 0
  m = X.shape[0]
  A = [0 for i in range(len(X))]
  cost = 0

  for i in range(len(X)):
    A[i] = float(sigmoid(w0 + w1*X.iloc[i,0] + w2*X.iloc[i,1] + w3*X.iloc[i,2] + w4*X.iloc[i,3] + w5*X.iloc[i,4] + w6*X.iloc[i,5] + w7*X.iloc[i,6]))
    cost = cost + compute_cost(A[i] , Y.iloc[i,0])
    dw0 = dw0 + (A[i] - Y.iloc[i,0])
    dw1 = dw1 + (A[i] - Y.iloc[i,0])*X.iloc[i,0]
    dw2 = dw2 + (A[i] - Y.iloc[i,0])*X.iloc[i,1]
    dw3 = dw3 + (A[i] - Y.iloc[i,0])*X.iloc[i,2]

    dw4 = dw4 + (A[i] - Y.iloc[i,0])*X.iloc[i,3]
    dw5 = dw5 + (A[i] - Y.iloc[i,0])*X.iloc[i,4]
    dw6 = dw6 + (A[i] - Y.iloc[i,0])*X.iloc[i,5]
    dw7 = dw7 + (A[i] - Y.iloc[i,0])*X.iloc[i,6]



  cost = (-1/m)*cost
  w0 = w0 - learning_rate * (dw0/m)
  w1 = w1 - learning_rate * (dw1/m)
  w2 = w2 - learning_rate * (dw2/m)
  w3 = w3 - learning_rate * (dw3/m)

  w4 = w4 - learning_rate * (dw4/m)
  w5 = w5 - learning_rate * (dw5/m)
  w6 = w6 - learning_rate * (dw6/m)
  w7 = w7 - learning_rate * (dw7/m)

  return w0,w1,w2,w3,w4,w5,w6,w7,cost
```

```
## logistic regression
def log_reg(X, Y, learning_rate = 0.01, num_iter = 1000):
  np.random.seed(7)
  w0 = float(np.random.randn(1,1))
  w1 = float(np.random.randn(1,1))
  w2 = float(np.random.randn(1,1))
  w3 = float(np.random.randn(1,1))
  w4 = float(np.random.randn(1,1))
  w5 = float(np.random.randn(1,1))
  w6 = float(np.random.randn(1,1))
  w7 = float(np.random.randn(1,1))

  cost_list = []

  for i in range(num_iter):
    w0, w1, w2, w3, w4, w5, w6, w7, cost = update_params(X, Y, w0, w1, w2, w3, w4, w5, w6, w7, learning_rate)
    cost_list.append(cost)

  return w0, w1, w2, w3, w4, w5, w6, w7, cost, cost_list  
```

```
def probs(X, w0, w1, w2, w3, w4, w5, w6, w7):
  A_test = [0 for i in range(len(X))]
  for i in range(len(X)):
    A_test[i] = (w0 + w1*X.iloc[i,0] + w2*X.iloc[i,1] + w3*X.iloc[i,2] + w4*X.iloc[i,3] + w5*X.iloc[i,4] + w6*X.iloc[i,5] + w7*X.iloc[i,6])
  return A_test
```

```
def predict(A1, A2, A3):
  Y_out = [0 for i in range(len(A1))]
  lst = [0 for i in range(len(A1))]
  for i in range(len(A1)):
    lst[i] = [A1[i], A2[i], A3[i]]
    Y_out[i] = lst[i].index(max(lst[i])) + 1
    
  Y_out = pd.DataFrame(Y_out)
  return Y_out
```

```
def confusion_matrix(Y, Yp):
  conf = np.zeros((3,3))
  for i in range(len(Y)):
    if(Y.iloc[i,0] == 1):
      if(Yp.iloc[i,0] == 1):
        conf[0][0] += 1
      elif(Yp.iloc[i,0] == 2):
        conf[0][1] += 1
      else:
        conf[0][2] += 1

    if(Y.iloc[i,0] == 2):
      if(Yp.iloc[i,0] == 1):
        conf[1][0] += 1
      elif(Yp.iloc[i,0] == 2):
        conf[1][1] += 1
      else:
        conf[1][2] += 1

    if(Y.iloc[i,0] == 3):
      if(Yp.iloc[i,0] == 1):
        conf[2][0] += 1
      elif(Yp.iloc[i,0] == 2):
        conf[2][1] += 1
      else:
        conf[2][2] += 1

  return conf
```

```
## k fold cross validation (k=5)
def k_cross_validation(X, Y, k):
  a = 0
  b = round(len(Y)/k)

  Acc_list_1 = []                                                       # list to append accuracy after each fold
  Acc_list_2 = []
  Acc_list_3 = []
  Acc_list_overall = []

  for i in range(k):
    #print("a and b is: %d, %d"%(a,b))
    if(i != k-1):                                                        # splitting the data into train and test set
      X_ts = X.iloc[a:b, :].reset_index(drop = True)
      Y_ts = Y.iloc[a:b, :].reset_index(drop = True)
      X_tr = pd.concat((X.iloc[:a,:], X.iloc[b:,:]), axis = 0).reset_index(drop = True)
      Y_tr = pd.concat((Y.iloc[:a,:], Y.iloc[b:,:]), axis = 0).reset_index(drop = True)

    else:                                                                # handling the last case (if number of samples is not divisible by k)
      X_ts = X.iloc[a:,:].reset_index(drop = True) 
      Y_ts = Y.iloc[a:,:].reset_index(drop = True)
      X_tr = X.iloc[:a,:].reset_index(drop = True)
      Y_tr = Y.iloc[:a,:].reset_index(drop = True)
    
    Y_tr.set_axis([0], axis = 1, inplace = True)                    # set column name to 0
    Y_ts.set_axis([0], axis = 1, inplace = True)

    Y_tr1 = pd.DataFrame((Y_tr[0] == 1).astype(int))              # label_1 = 1 and label_2 and label_3 = 0
    Y_tr2 = pd.DataFrame((Y_tr[0] == 2).astype(int))              # label_2 = 1 and label_1 and label_3 = 0
    Y_tr3 = pd.DataFrame((Y_tr[0] == 3).astype(int))              # label_3 = 1 and label_2 and label_1 = 0
    
    min_X = list(X_tr.min(axis = 0))                              # min value of features of training data
    max_X = list(X_tr.max(axis = 0))                              # max value of features of training data

    X_tr_norm = normalize(X_tr, min_X, max_X)                     # normalising training and test data
    X_ts_norm = normalize(X_ts, min_X, max_X)

    ### logistic regression one vs all ###
    w0_1, w1_1, w2_1, w3_1, w4_1, w5_1, w6_1, w7_1, cost_1, cost_l_1 = log_reg(X_tr_norm, Y_tr1, learning_rate = 0.05, num_iter = 500)
    w0_2, w1_2, w2_2, w3_2, w4_2, w5_2, w6_2, w7_2, cost_2, cost_l_2 = log_reg(X_tr_norm, Y_tr2, learning_rate = 0.02, num_iter = 500)
    w0_3, w1_3, w2_3, w3_3, w4_3, w5_3, w6_3, w7_3, cost_3, cost_l_3 = log_reg(X_tr_norm, Y_tr3, learning_rate = 0.07, num_iter = 500)

    ### computing Acivations ###
    A_ts1 = probs(X_ts_norm, w0_1, w1_1, w2_1, w3_1, w4_1, w5_1, w6_1, w7_1)
    A_ts2 = probs(X_ts_norm, w0_2, w1_2, w2_2, w3_2, w4_2, w5_2, w6_2, w7_2)
    A_ts3 = probs(X_ts_norm, w0_3, w1_3, w2_3, w3_3, w4_3, w5_3, w6_3, w7_3)

    ### Predicting output ####
    Y_predicted = predict(A_ts1, A_ts2, A_ts3)

    ### confusion_matrix ####
    conf_mat = confusion_matrix(Y_ts, Y_predicted)

    ### calculating accuracy of each label as well as overall accuracy
    Acc_1 = conf_mat[0][0]/np.sum(conf_mat[0], axis = 0)
    Acc_2 = conf_mat[1][1]/np.sum(conf_mat[1], axis = 0)
    Acc_3 = conf_mat[2][2]/np.sum(conf_mat[2], axis = 0)
    Acc_overall = (conf_mat[0][0]+conf_mat[1][1]+conf_mat[2][2])/np.sum(conf_mat)

    print("Acc_1 in iter %d is %.2f" %(i, Acc_1*100)+ " %")
    print("Acc_2 in iter %d is %.2f" %(i, Acc_2*100)+ " %")
    print("Acc_3 in iter %d is %.2f" %(i, Acc_3*100)+ " %")
    print("Acc_overall in iter %d is %.2f" %(i, Acc_overall*100)+ " %")
    print()
    
    ### appending accuracies in the list
    #Acc_list_1.append(Acc_1)
    #Acc_list_2.append(Acc_2)
    #Acc_list_3.append(Acc_3)
    #Acc_list_overall.append(Acc_overall)

    a = a + round(len(Y)/k)
    b = b + round(len(Y)/k)

  #return Acc_list_1, Acc_list_2, Acc_list_3, Acc_list_overall
```
