```
import pandas as pd 
import numpy as np
```

```
data = pd.read_excel('data4.xlsx' ,header = None)
data.head()
```

```
data_shuffled = data.sample(frac = 1, random_state = 7).reset_index(drop = True)
data_shuffled
```

```
##### separate input(X) and output(Y) #####

X = data_shuffled.iloc[:,:-1]
Y = data_shuffled.iloc[:,-1:]
#print("First 5 rows of X is: ")
#print(X.head())
#print()
#print("First 5 rows of Y is: ")
#print(Y.head())
```

```
###### splitting the X and Y in train and test data #######

def split(X, Y, split_size = 0.6):
  train_size = int(len(X) * split_size)
  X_tr = X.iloc[:train_size, :]
  X_ts = X.iloc[train_size:, :]
  Y_tr = Y.iloc[:train_size, :]
  Y_ts = Y.iloc[train_size:, :]

  return X_tr, X_ts, Y_tr, Y_ts
```

```
X_train, X_test, Y_train, Y_test = split(X, Y, split_size = 0.6)
X_test = X_test.reset_index(drop = True)                             # to set index
Y_test = Y_test.reset_index(drop = True)                             # to set index
Y_train.set_axis([0], axis = 1, inplace = True)                      # to set column name to 0
Y_test.set_axis([0], axis = 1, inplace = True)                       # to set column name to 0
```

```
print("X_train shape is: ", X_train.shape)
print("Y_train shape is: ", Y_train.shape)
print("X_test shape is: ", X_test.shape)
print("Y_test shape is: ", Y_test.shape)
```

```
min_X = list(X_train.min(axis = 0))
max_X = list(X_train.max(axis = 0))

def normalize(D, min_features, max_features):
  Z = pd.DataFrame()
  for i in range(D.shape[1]):
    Z[i] = (D.iloc[:,i] - min_X[i])/(max_X[i] - min_X[i])
  return Z
```

```
X_train_norm = normalize(X_train, min_X, max_X)
```

```
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```
def compute_cost(A, Y):
  cost_val = (Y*np.log(A)) + ((1-Y)*np.log(1-A))
  return cost_val 
```

```
def update_params(X, Y, w0, w1, w2, w3, w4, w5, w6, w7, learning_rate = 0.01):
  dw0 = dw1 = dw2 = dw3 = dw4 = dw5 = dw6 = dw7 = 0
  m = X.shape[0]
  A = [0 for i in range(len(X))]
  cost = 0

  for i in range(len(X)):
    A[i] = float(sigmoid(w0 + w1*X.iloc[i,0] + w2*X.iloc[i,1] + w3*X.iloc[i,2] + w4*X.iloc[i,3] + w5*X.iloc[i,4] + w6*X.iloc[i,5] + w7*X.iloc[i,6]))
    cost = cost + compute_cost(A[i] , Y.iloc[i,0])
    dw0 = dw0 + (A[i] - Y.iloc[i,0])
    dw1 = dw1 + (A[i] - Y.iloc[i,0])*X.iloc[i,0]
    dw2 = dw2 + (A[i] - Y.iloc[i,0])*X.iloc[i,1]
    dw3 = dw3 + (A[i] - Y.iloc[i,0])*X.iloc[i,2]

    dw4 = dw4 + (A[i] - Y.iloc[i,0])*X.iloc[i,3]
    dw5 = dw5 + (A[i] - Y.iloc[i,0])*X.iloc[i,4]
    dw6 = dw6 + (A[i] - Y.iloc[i,0])*X.iloc[i,5]
    dw7 = dw7 + (A[i] - Y.iloc[i,0])*X.iloc[i,6]



  cost = (-1/m)*cost
  w0 = w0 - learning_rate * (dw0/m)
  w1 = w1 - learning_rate * (dw1/m)
  w2 = w2 - learning_rate * (dw2/m)
  w3 = w3 - learning_rate * (dw3/m)

  w4 = w4 - learning_rate * (dw4/m)
  w5 = w5 - learning_rate * (dw5/m)
  w6 = w6 - learning_rate * (dw6/m)
  w7 = w7 - learning_rate * (dw7/m)

  return w0,w1,w2,w3,w4,w5,w6,w7,cost
```

```
def log_reg(X, Y, learning_rate = 0.01, num_iter = 1000):
  np.random.seed(7)
  w0 = float(np.random.randn(1,1))
  w1 = float(np.random.randn(1,1))
  w2 = float(np.random.randn(1,1))
  w3 = float(np.random.randn(1,1))
  w4 = float(np.random.randn(1,1))
  w5 = float(np.random.randn(1,1))
  w6 = float(np.random.randn(1,1))
  w7 = float(np.random.randn(1,1))

  cost_list = []

  for i in range(num_iter):
    w0, w1, w2, w3, w4, w5, w6, w7, cost = update_params(X, Y, w0, w1, w2, w3, w4, w5, w6, w7, learning_rate)
    cost_list.append(cost)

  return w0, w1, w2, w3, w4, w5, w6, w7, cost, cost_list  
```

```
X_test_norm = normalize(X_test, min_X, max_X)
```

```
def probs(X, w0, w1, w2, w3, w4, w5, w6, w7):
  A_test = [0 for i in range(len(X))]
  Y_pred = [0 for i in range(len(X))]
  for i in range(len(X)):
    A_test[i] = float(sigmoid(w0 + w1*X.iloc[i,0] + w2*X.iloc[i,1] + w3*X.iloc[i,2] + w4*X.iloc[i,3] + w5*X.iloc[i,4] + w6*X.iloc[i,5] + w7*X.iloc[i,6]))
    Y_pred[i] = 1 if (A_test[i] >= 0.5) else 0 
  return Y_pred
```

```
Y_test_1_2 = probs(X_test_norm, w0_12, w1_12, w2_12, w3_12, w4_12, w5_12, w6_12, w7_12)
Y_test_1_3 = probs(X_test_norm, w0_13, w1_13, w2_13, w3_13, w4_13, w5_13, w6_13, w7_13)
Y_test_2_3 = probs(X_test_norm, w0_23, w1_23, w2_23, w3_23, w4_23, w5_23, w6_23, w7_23)
```

```
def predict(Y1, Y2, Y3):
  Y_pred = [0 for i in range(len(Y1))]
  for i in range(len(Y1)):
    if(Y1[i] == 1 and Y2[i] == 1):
      Y_pred[i] = 1
    elif(Y1[i] == 0 and Y3[i] == 1):
      Y_pred[i] = 2
    elif(Y2[i] == 0 and Y3[i] == 0):
      Y_pred[i] = 3
    else:
      Y_pred[i] = 1
  
  return Y_pred
```

```
Y_predicted = predict(Y_test_1_2, Y_test_1_3, Y_test_2_3)
```

```
df_to_compare = Y_test.copy()
df_to_compare[1] = Y_predicted
df_to_compare.columns = ['Y_test', 'Y_predicted']
```

```
confusion_matrix = pd.crosstab(df_to_compare['Y_test'], df_to_compare['Y_predicted'], rownames = ['actual'], colnames = ['predicted'])
confusion_matrix
```

```
Acc_1 = confusion_matrix.iloc[0,0]/np.sum(confusion_matrix.iloc[0,:])
Acc_2 = confusion_matrix.iloc[1,1]/np.sum(confusion_matrix.iloc[1,:])
Acc_3 = confusion_matrix.iloc[2,2]/np.sum(confusion_matrix.iloc[2,:])
Acc_overall = (confusion_matrix.iloc[0,0]+confusion_matrix.iloc[1,1]+confusion_matrix.iloc[2,2])/np.sum(np.sum(confusion_matrix))
```

```
print("Accuracy of 1 is: %f" %(Acc_1*100)+' %')
print("Accuracy of 2 is: %f" %(Acc_2*100)+' %')
print("Accuracy of 3 is: %f" %(Acc_3*100)+' %')
print("Overall Accuracy is: %f" %(Acc_overall*100)+' %')
```
